```python
import pandas as pd
import numpy as np
import keras
from keras.layers import *
from keras.models import Model
from keras import optimizers
import random
```

```python
from nltk.stem import WordNetLemmatizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.optimizers import SGD
import tensorflow as tf
```

```python
train_og = pd.read_csv("/content/drive/MyDrive/train.csv")
train = train_og.copy()
```

```python
train.head(10)
```

```python
X_train=train.iloc[:,1:].astype('float32')
X_train=X_train/255
```

```python
y_train=train_og.iloc[:, 0]
y_traincpy  = y_train.copy()
```

```python
y_train
```

```python
num_digits = 10
y_train = keras.utils.to_categorical(y_train, num_digits)
```

```python
y_train
```

```python
input_dim = 784
output_units = 10

model = Sequential()
model.add(Dense(784, input_dim=input_dim))
model.add(Dense(128,  activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(32,  activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(output_units, activation='softmax'))

training_epochs = 10


model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(X_train, y_train,
                      epochs = training_epochs,
                      validation_split=0.2, shuffle=True)
```

```python
from sklearn.metrics import f1_score
```

```python
preds = model.predict(X_train)
preds = np.argmax(preds, axis=1)
preds
```

```python
f1_score(y_traincpy, preds, average='macro')
```
